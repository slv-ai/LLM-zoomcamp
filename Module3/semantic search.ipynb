1.DATA PREPARATION

```python
import json

with open('documents.json', 'rt') as f_in:
    docs_raw = json.load(f_in)
```

```python
documents = []

for course_dict in docs_raw:
    for doc in course_dict['documents']:
        doc['course'] = course_dict['course']
        documents.append(doc)
```

```python
documents[0]
```

```python
#2.EMBEDDINGS using pretrained models
```

```python
pip install sentence_transformers==2.7.0
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-mpnet-base-v2")
```

```python
len(model.encode("this is a vector"))
```

```python
#create a dense vector using pretrained models
answers=[]
for doc in documents:
    doc["text_vector"]=model.encode(doc["text"]).tolist()
    answers.append(doc)
```

```python
answers[0]
```

```python
#3.setup elastic search connection
```

```python
pip install elasticsearch
```

```python
from elasticsearch import Elasticsearch
es_client = Elasticsearch('http://localhost:9200') 

es_client.info()
```

```python
#4.create mappings and index
```

```python
index_settings = {
    "settings": {
        "number_of_shards": 1,
        "number_of_replicas": 0
    },
    "mappings": {
        "properties": {
            "text": {"type": "text"},
            "section": {"type": "text"},
            "question": {"type": "text"},
            "course": {"type": "keyword"} ,
            "text_vector": {"type": "dense_vector", "dims": 768, "index": True, "similarity": "cosine"},
        }
    }
}
```

```python
index_name = "course-questions"

es_client.indices.delete(index=index_name, ignore_unavailable=True)
es_client.indices.create(index=index_name, body=index_settings)
```

```python
#5.adding the documents to elastic search index
```

```python
for doc in answers:
    try:
        es_client.index(index=index_name,document=doc)
    except exception as e:
        print(e)
```

```python
#6.create query and Keyword Search
```

```python
search_term = "windows or mac?"
vector_search_term = model.encode(search_term)
```

```python
query={
    "field":"text_vector",
    "query_vector": vector_search_term,
    "k":5,
    "num_candidates": 10000,
}
```

```python
result=es_client.search(index=index_name,knn=query,source=["text","section","question","course"])
result['hits']['hits']
```
