```python
from elasticsearch import Elasticsearch
```

```python
es_client=Elasticsearch('http://localhost:9200')
index_settings={
    "settings":{
        "number_of_shards":1,
        "number_of_replicas":0
    },
    "mappings":{
        "properties":{
            "resume_id":{"type":"keyword"},
            "Resume":{"type":"text"},
            "Resume_vector":{
                "type":"dense_vector",
                "dims": 768,
                "index": True,
                "similarity": "cosine"
            },
        }
    }
}
index_name="candiate_resume"
es_client.indices.delete(index=index_name,ignore_unavailable=True)
es_client.indices.create(index=index_name,body=index_settings)


            
```

```python
from tqdm.auto import tqdm
import numpy as np

for doc in tqdm(docs):
    try:
        resume_id = doc['resume_id']
        Resume = doc['Resume']
        vector = model.encode(Resume)

        # Convert numpy array to list if necessary
        if isinstance(vector, np.ndarray):
            vector = vector.tolist()

        # Ensure vector length matches the expected dimension
        if len(vector) != 768:
            raise ValueError(f"Vector length mismatch: expected 768, got {len(vector)}")

        doc['Resume_vector'] = vector

        # Print document to verify structure
        #print(f"Indexing document: {doc}")

        es_client.index(index=index_name, document=doc)
        #print(response)
    except Exception as e:
        print(f"Error indexing document {resume_id}: {e}")
```

```python
def elastic_search_knn(field,vector):
    knn={
        "field":field,
        "query_vector":vector,
        "k":5,
        "num_candidates" : 1000
    }
    search_query={
        "knn":knn,
        "_source":["resume_id","Resume"]
    }
    es_results=es_client.search(
        index=index_name,
        body=search_query
    )
    sorted_resume=[]
    for hit in es_results['hits']['hits']:
        sorted_resume.append(hit['_source'])
    return sorted_resume
    
    
```

```python
def job_describ_vector(q):
    job_describ=q['job_describ']
    
    question_vect=model.encode(job_describ)

    return elastic_search_knn('Resume_vector',question_vect)
```
